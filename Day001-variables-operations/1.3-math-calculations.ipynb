```python
"""
ЗАДАЧА 3: Математические вычисления
реализация первых полезных калькуляторов
"""

print(" Математические калькуляторы")
print("=" * 35)

print("\n Калькулятор ИМТ:")
weight = float(input("Ваш вес (кг): "))
height = float(input("Ваш рост (м): "))
bmi = weight / (height ** 2)
print(f"Ваш ИМТ: {bmi:.1f}")

print("\n Конвертер валют:")
rub_amount = float(input("Сумма в рублях: "))
usb_rate = 0.012
usd_amount = rub_amount * usb_rate
print(f"{rub_amount} RUB = {usd_amount:.2f} USD")

print("\n КАЛЬКУЛЯТОР СКИДОК:")
original_price = float(input("Исходная цена: "))
discount_percent = float(input("Процент скидки: "))
discount_amount = original_price * (discount_percent / 100)
final_price = original_price - discount_amount
print(f"Скидка: {discount_amount:.2f} руб")
print(f"Итоговая цена: {final_price:.2f} руб")
```
