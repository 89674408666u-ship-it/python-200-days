```python
"""–ó–ê–î–ê–ß–ê 1: –ê–Ω–∫–µ—Ç–∞ –±—É–¥—É—â–µ–≥–æ —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–∞
–¶–µ–ª—å: –ö–æ–º–ø–ª–µ–∫—Å–Ω–∞—è —Ä–∞–±–æ—Ç–∞ —Å –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏, –≤–≤–æ–¥–æ–º –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ–º"""

print("üöÄ –ê–ù–ö–ï–¢–ê –ë–£–î–£–©–ï–ì–û –†–ê–ó–†–ê–ë–û–¢–ß–ò–ö–ê üöÄ")
print("=" * 50)

# –†–∞–∑–¥–µ–ª 1: –õ–∏—á–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
print('\nüìù –õ–ò–ß–ù–ê–Ø –ò–ù–§–û–†–ú–ê–¶–ò–Ø:')
name = input("‚Ä¢ –í–∞—à–µ –∏–º—è: ")
age = int(input("‚Ä¢ –í–∞—à –≤–æ–∑—Ä–∞—Å—Ç: "))
city = input("‚Ä¢ –í–∞—à –≥–æ—Ä–æ–¥: ")

# –†–∞–∑–¥–µ–ª 2: –ú–æ—Ç–∏–≤–∞—Ü–∏—è –∏ —Ü–µ–ª–∏
print('\nüéØ –ú–û–¢–ò–í–ê–¶–ò–Ø –ò –¶–ï–õ–ò:')
reason = input("‚Ä¢ –ü–æ—á–µ–º—É –≤—ã —Ö–æ—Ç–∏—Ç–µ —Å—Ç–∞—Ç—å —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–æ–º? ")
goal = input("‚Ä¢ –ö–∞–∫–æ–π –ø—Ä–æ–µ–∫—Ç –º–µ—á—Ç–∞–µ—Ç–µ —Å–æ–∑–¥–∞—Ç—å? ")
deadline = input("‚Ä¢ –ö–æ–≥–¥–∞ –ø–ª–∞–Ω–∏—Ä—É–µ—Ç–µ –¥–æ—Å—Ç–∏—á—å —Ü–µ–ª–∏? ")

# –†–∞–∑–¥–µ–ª 3: –¢–µ—Ö–Ω–∏—á–µ—Å–∫–∏–µ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è
print("\nüíª –¢–ï–•–ù–ò–ß–ï–°–ö–ò–ï –ü–†–ï–î–ü–û–ß–¢–ï–ù–ò–Ø:")
experience = input("‚Ä¢ –ï—Å—Ç—å –ª–∏ –æ–ø—ã—Ç –≤ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–∏? (–¥–∞/–Ω–µ—Ç): ")
preferred_os = input("‚Ä¢ –ö–∞–∫–∞—è –û–° –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è? (Windows/Mac/Linux): ")
hours_per_day = int(input("‚Ä¢ –°–∫–æ–ª—å–∫–æ —á–∞—Å–æ–≤ –≤ –¥–µ–Ω—å –≥–æ—Ç–æ–≤—ã —É—á–∏—Ç—å—Å—è? "))  # ‚Üê –°–û–•–†–ê–ù–Ø–ï–ú –∑–Ω–∞—á–µ–Ω–∏–µ!

# –†–∞—Å—á–µ—Ç—ã –∏ –∞–Ω–∞–ª–∏—Ç–∏–∫–∞
print("\nüìä –ê–ù–ê–õ–ò–¢–ò–ö–ê –í–ê–®–ï–ì–û –ü–£–¢–ò:")
days_to_goal = 200
total_hours = days_to_goal * hours_per_day  # ‚Üê –ò—Å–ø–æ–ª—å–∑—É–µ–º –í–í–ï–î–ï–ù–ù–û–ï –∑–Ω–∞—á–µ–Ω–∏–µ!
completion_year = 2026  # ‚Üê –§–∏–∫—Å–∏—Ä—É–µ–º –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≥–æ–¥

print(f"‚Ä¢ –î–æ —Ü–µ–ª–∏ –æ—Å—Ç–∞–ª–æ—Å—å: {days_to_goal} –¥–Ω–µ–π")
print(f"‚Ä¢ –í—Å–µ–≥–æ —á–∞—Å–æ–≤ –æ–±—É—á–µ–Ω–∏—è: {total_hours} —á–∞—Å–æ–≤")
print(f"‚Ä¢ –î–∞—Ç–∞ –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è —Ü–µ–ª–∏: –∏—é–ª—å {completion_year} –≥–æ–¥–∞")
print(f"‚Ä¢ –í—ã–¥–µ–ª—è—è {hours_per_day} —á–∞—Å–æ–≤ –≤ –¥–µ–Ω—å, —Ç—ã –¥–æ—Å—Ç–∏–≥–Ω–µ—à—å —É—Å–ø–µ—Ö–∞!")

# –ú–æ—Ç–∏–≤–∞—Ü–∏–æ–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
print("\nüî• –í–ê–® –ü–£–¢–¨ –†–ê–ó–†–ê–ë–û–¢–ß–ò–ö–ê:")
print(f"–ü—Ä–∏–≤–µ—Ç—Å—Ç–≤—É—é, {name}! –ò–∑ {city} –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Ç–≤–æ–µ –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏–µ –≤ IT.")
print(f"–¢–≤–æ—è —Ü–µ–ª—å '{goal}' - —ç—Ç–æ –∞–º–±–∏—Ü–∏–æ–∑–Ω–æ –∏ –¥–æ—Å—Ç–∏–∂–∏–º–æ!")

if experience.lower() == "–Ω–µ—Ç":
    print("üéâ –¢—ã –Ω–∞—á–∏–Ω–∞–µ—à—å —Å —á–∏—Å—Ç–æ–≥–æ –ª–∏—Å—Ç–∞ - —ç—Ç–æ —Ç–≤–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ!")
else:
    print("üí™ –û–ø—ã—Ç –ø–æ–º–æ–∂–µ—Ç —Ç–µ–±–µ –¥–≤–∏–≥–∞—Ç—å—Å—è –±—ã—Å—Ç—Ä–µ–µ!")

print("\n" + "=" * 50)
print("–ü–û–ú–ù–ò: –ö–∞–∂–¥—ã–π –≤–µ–ª–∏–∫–∏–π —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫ –∫–æ–≥–¥–∞-—Ç–æ –Ω–∞—á–∏–Ω–∞–ª —Å –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–∏ –∫–æ–¥–∞!")
```
