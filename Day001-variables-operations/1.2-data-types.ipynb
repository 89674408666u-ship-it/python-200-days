```python
"""Задача 2: Работа с разными типами данных
Освоить преобразование между типами и базовые операции
"""

print("Работа с разными типами данных")
print("=" * 40)

num1 = 25
num2 = 7
# Часть 1: Числовые типы
print(f"Исходные числа{num1} и {num2}")
print(f"Сумма(+):{num1+num2}")
print(f"Разность(-):{num1-num2}")
print(f"Произведение(*):{num1*num2}")
print(f"Обычное деление(/):{num1/num2}")
print(f"Целочисленное деление(//):{num1//num2}")
print(f"Остаток от деления(%):{num1%num2}")
print(f"Возведение в степень(**):{num1**num2}")

#Преобразование типов
print("\n ПРЕОБРАЗОВАНИЕ ТИПОВ:")
price = "1500"
quantity = "3"
total_price = int(price) * int(quantity)
print(f"Цена: {price} (тип: {type(price)})")
print(f"Количество:{quantity} (тип:{type(quantity)})")
print(f"Общая стоимость: {total_price} (тип: {type(total_price)})")

# Часть 3: Булевы значения
print("\n БУЛЕВА ЛОГИКА:")
is_weekend = True
has_homework = False
can_relax = is_weekend and not has_homework
print(f"Выходной: {is_weekend}")
print(f"Есть домашняя работа: {has_homework}")
print(f"Можно отдыхать: {can_relax}")



```
